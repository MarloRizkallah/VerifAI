```python
#Only run 1 time.
#Here we iterate through the wiki-pages, then through each line in the file(extract the page id from each json object).
#so each page is jsonl, which has lines of json objects, and each json object represents 1 wiki-pedia page, so we're building
#a lookup for the page id where we can extract evidence for claims in the following steps.
import os
import json
from tqdm import tqdm

WIKI_FOLDER = r"Data\Wikipedia dump\wiki-pages"

page_to_file = {}

files = os.listdir(WIKI_FOLDER)

for filename in tqdm(files):
    file_path = os.path.join(WIKI_FOLDER, filename)

    with open(file_path, "r", encoding="utf-8") as f:
        for line in f:
            if not line.strip():
                continue

            data = json.loads(line)
            page_id = data["id"]

            if page_id:  # ignore empty id
                page_to_file[page_id] = filename

print("Total indexed pages:", len(page_to_file))
```

```python
import os
import json

WIKI_FOLDER = r"Data\Wikipedia dump\wiki-pages"

wiki_cache = {}

def get_sentence_from_wiki(page_name, sentence_id):
    # Use cache if already loaded
    if page_name in wiki_cache:
        return wiki_cache[page_name].get(sentence_id)

    # Find which file contains this page
    filename = page_to_file.get(page_name)

    if not filename:
        return None

    file_path = os.path.join(WIKI_FOLDER, filename)

    # Open only that file
    with open(file_path, "r", encoding="utf-8") as f:
        for line in f:
            if not line.strip():
                continue

            data = json.loads(line)

            if data["id"] == page_name:
                sentences = {}

                for l in data["lines"].split("\n"):
                    if not l.strip():
                        continue

                    parts = l.split("\t")
                    if len(parts) >= 2:
                        sid = int(parts[0])
                        text = parts[1]
                        sentences[sid] = text

                # Cache the page
                wiki_cache[page_name] = sentences

                return sentences.get(sentence_id)

    return None
```

```python
# -------------------------------------------------------------
# STEP 1: Extract claims labeled "NOT ENOUGH INFO" from clean_pairs
# These claims currently have no evidence text.
# We will generate synthetic evidence for them (random + BM25).
# -------------------------------------------------------------

nei_claims = [p for p in clean_pairs if p["label"] == "NOT ENOUGH INFO"]

print("Total NEI claims:", len(nei_claims))
```

```python
# -------------------------------------------------------------
# STEP 2: Build a subset of Wikipedia sentences from the dump
#
# Purpose:
#   Extract a large collection of sentences that can be used
#   as negative evidence for NOT ENOUGH INFO (NEI) examples.
#
# Strategy:
#   - Scan the Wikipedia dump files sequentially
#   - Extract sentences from the "lines" field
#   - Stop once we reach the desired subset size
#
# Using a subset avoids loading the entire Wikipedia dump
# into memory while still providing enough diversity.
# -------------------------------------------------------------

import os
import json

WIKI_FOLDER = r"Data\Wikipedia dump\wiki-pages"

MAX_SENTENCES = 1_000_000   # desired subset size

subset_wiki_sentences = []

for filename in os.listdir(WIKI_FOLDER):

    file_path = os.path.join(WIKI_FOLDER, filename)

    with open(file_path, "r", encoding="utf-8") as f:
        for line in f:

            # Parse each Wikipedia page JSON
            data = json.loads(line)

            # Extract sentences from the "lines" field
            for l in data["lines"].split("\n"):

                if not l.strip():
                    continue

                parts = l.split("\t")

                if len(parts) >= 2:
                    sentence = parts[1].strip()

                    # Filter very short sentences
                    if len(sentence) > 20:
                        subset_wiki_sentences.append(sentence)

                # Stop when subset size is reached
                if len(subset_wiki_sentences) >= MAX_SENTENCES:
                    break

            if len(subset_wiki_sentences) >= MAX_SENTENCES:
                break

    if len(subset_wiki_sentences) >= MAX_SENTENCES:
        break


print(f"Sentence pool created: {len(subset_wiki_sentences):,} sentences")
```

```python
# -------------------------------------------------------------
# STEP 3: Build BM25 index over the subset of Wikipedia sentences
#
# Functional purpose:
#   - Construct a lexical retrieval engine for candidate evidence
#   - Treat each Wikipedia sentence as a searchable document
#   - Pre-tokenize sentences for faster retrieval during NEI generation
#
# Why BM25 is used:
#   BM25 retrieves sentences that share lexical overlap with claims.
#   These candidates are later reranked using SBERT to obtain
#   semantically related but non-verifying evidence.
#
# Input expected:
#   subset_wiki_sentences -> list[str]
#       A subset (~1M) of Wikipedia sentences extracted earlier.
#
# Output produced:
#   bm25 -> BM25Okapi retrieval object
# -------------------------------------------------------------

from rank_bm25 import BM25Okapi

print("Building BM25 index...")

# Pre-tokenize each sentence once for faster retrieval
tokenized_corpus = [sentence.lower().split() for sentence in subset_wiki_sentences]

# Initialize BM25 retrieval model
bm25 = BM25Okapi(tokenized_corpus)

print("BM25 index ready.")
print("Total indexed sentences:", len(subset_wiki_sentences))
```

```python
# -------------------------------------------------------------
# STEP 4: Generate NOT ENOUGH INFO (NEI) training pairs
#
# Functional purpose:
#   Construct NEI examples for the NLI verifier training dataset.
#
# Strategy used:
#   1. 33% Random negatives
#      → Teach the model to recognize completely irrelevant evidence.
#
#   2. 33% BM25 + SBERT (mid similarity)
#      → Teach the model that topically related evidence can still be insufficient.
#
#   3. 33% BM25 + SBERT (high similarity)
#      → Hard negatives: semantically very close but still not verifying the claim.
#
# Retrieval pipeline:
#   Claim → BM25 retrieves lexical candidates → SBERT reranks semantically
#
# Additional design choices:
#   - SBERT encodes claim + candidates together (batch) for efficiency
#   - Progress checkpoints save intermediate results every 2000 samples
#   - Metadata fields (source, similarity) are stored for later analysis
#
# Output produced:
#   nei_pairs → list of NEI training examples
#   nei_pairs_checkpoint.jsonl → periodic crash-safe saves
#   nei_pairs_sbert.jsonl → final NEI dataset
# -------------------------------------------------------------

from sentence_transformers import SentenceTransformer, util
from tqdm import tqdm
import random
import numpy as np
import jsonlines

# Fix random seed for reproducibility (important for experiments/papers)
random.seed(42)

# Load SBERT semantic similarity model
print("Loading SBERT model...")
sbert = SentenceTransformer("all-MiniLM-L6-v2")

# Container for generated NEI pairs
nei_pairs = []

# Counters for monitoring dataset distribution
random_count = 0
mid_count = 0
high_count = 0


# -------------------------------------------------------------
# Main NEI generation loop
# -------------------------------------------------------------
for pair in tqdm(nei_claims):

    # Extract claim text
    claim = pair["claim"]

    # Tokenize claim for BM25 retrieval
    tokens = claim.lower().split()

    # Random number determines NEI generation strategy
    r = random.random()

    # ---------------------------------------------------------
    # CASE 1: Random negative evidence (≈33%)
    # ---------------------------------------------------------
    if r < 0.33:

        # Select completely random Wikipedia sentence
        evidence = random.choice(subset_wiki_sentences)

        nei_pairs.append({
            "claim": claim,
            "evidence": evidence,
            "label": "NOT ENOUGH INFO",
            "source": "random",
            "similarity": None
        })

        random_count += 1


    # ---------------------------------------------------------
    # CASE 2 & 3: BM25 retrieval + SBERT semantic reranking
    # ---------------------------------------------------------
    else:

        # Retrieve top lexical candidates using BM25
        candidates = bm25.get_top_n(tokens, subset_wiki_sentences, n=20)

        # Encode claim and candidates together for efficient batching
        sentences = [claim] + candidates
        embeddings = sbert.encode(sentences, batch_size=32, convert_to_tensor=True)

        claim_emb = embeddings[0]
        cand_embs = embeddings[1:]

        # Compute cosine similarity between claim and candidate sentences
        scores = util.cos_sim(claim_emb, cand_embs)[0].cpu().numpy()

        # Sort candidates by semantic similarity
        sorted_idx = np.argsort(scores)[::-1]


        # -----------------------------------------------------
        # CASE 2: Mid similarity semantic negatives (≈33%)
        # -----------------------------------------------------
        if r < 0.66:

            mid_idx = sorted_idx[len(sorted_idx)//2]

            evidence = candidates[mid_idx]
            similarity = float(scores[mid_idx])

            source = "bm25_sbert_mid"
            mid_count += 1


        # -----------------------------------------------------
        # CASE 3: High similarity hard negatives (≈33%)
        # -----------------------------------------------------
        else:

            high_idx = sorted_idx[0]

            evidence = candidates[high_idx]
            similarity = float(scores[high_idx])

            source = "bm25_sbert_high"
            high_count += 1


        # Store generated NEI pair
        nei_pairs.append({
            "claim": claim,
            "evidence": evidence,
            "label": "NOT ENOUGH INFO",
            "source": source,
            "similarity": similarity
        })


    # ---------------------------------------------------------
    # Crash-safe checkpoint saving
    # ---------------------------------------------------------
    # Every 2000 generated pairs we write current results to disk.
    # This prevents losing hours of computation if the notebook
    # crashes or the IDE stops responding.
    if len(nei_pairs) % 2000 == 0:
        with jsonlines.open("nei_pairs_checkpoint.jsonl", "w") as writer:
            writer.write_all(nei_pairs)
        print("Checkpoint saved:", len(nei_pairs))


# -------------------------------------------------------------
# Final dataset saving
# -------------------------------------------------------------
# Save the completed NEI dataset after generation finishes.
with jsonlines.open("nei_pairs_sbert.jsonl", "w") as writer:
    writer.write_all(nei_pairs)


# -------------------------------------------------------------
# Summary statistics
# -------------------------------------------------------------
print("NEI pairs generated:", len(nei_pairs))
print("Random:", random_count)
print("Mid similarity:", mid_count)
print("High similarity:", high_count)

print("Final NEI dataset saved to: nei_pairs_sbert.jsonl")
```

```python
import json
import re
from collections import Counter

def clean_artifacts(text):
    # Replace Wikipedia bracket encodings with actual punctuation
    text = text.replace("-LRB-", "(").replace("-RRB-", ")")
    text = text.replace("-LSB-", "[").replace("-RSB-", "]")
    text = text.replace("-LCB-", "{").replace("-RCB-", "}")
    # Clean up extra whitespace left after replacements
    text = re.sub(r'\s+', ' ', text).strip()
    return text

def remove_duplicates(pairs):
    # Remove pairs where both claim AND evidence are identical
    # Same claim with different evidence is kept (that's valid)
    seen = set()
    deduped = []
    for p in pairs:
        key = (p["claim"], p["evidence"])
        if key not in seen:
            seen.add(key)
            deduped.append(p)
    return deduped

def filter_short_evidence(pairs, min_words=4):
    # Remove evidence strings shorter than min_words
    # Threshold is 4 because anything below is a fragment, not a sentence
    return [p for p in pairs if len(p["evidence"].split()) >= min_words]

def apply_cleaning(pairs, label_name, min_words=4):
    print(f"=== Cleaning {label_name} ===")
    print(f"  Before:               {len(pairs)}")

    pairs = remove_duplicates(pairs)
    print(f"  After dedup:          {len(pairs)}")

    pairs = filter_short_evidence(pairs, min_words)
    print(f"  After short filter:   {len(pairs)}")

    pairs = [
        {**p,
         "claim":    clean_artifacts(p["claim"]),
         "evidence": clean_artifacts(p["evidence"])}
        for p in pairs
    ]

    # Second dedup pass to catch duplicates created by artifact cleaning
    pairs = remove_duplicates(pairs)
    print(f"  After post-clean dedup: {len(pairs)}")
    print()

    return pairs
```

```python
# Run cleaning on supports using the unified function
# (replaces the manual steps we did earlier)
supports = []
with open("supports_dataset.jsonl", "r", encoding="utf-8") as f:
    for line in f:
        supports.append(json.loads(line))

# Run full cleaning pipeline
supports_final = apply_cleaning(supports, "SUPPORTS")

# Save
with open("supports_final.jsonl", "w", encoding="utf-8") as f:
    for row in supports_final:
        f.write(json.dumps(row) + "\n")

print(f"Saved: supports_final.jsonl")
print(f"Final supports count: {len(supports_final)}")
```

```python
# Load refutes
refutes = []
with open("refutes_dataset.jsonl", "r", encoding="utf-8") as f:
    for line in f:
        refutes.append(json.loads(line))

# Run full cleaning pipeline
refutes_final = apply_cleaning(refutes, "REFUTES")

# Save
with open("refutes_final.jsonl", "w", encoding="utf-8") as f:
    for row in refutes_final:
        f.write(json.dumps(row) + "\n")

print(f"Saved: refutes_final.jsonl")
print(f"Final refutes count: {len(refutes_final)}")
```

```python
# ============================================================
# NEI Dataset Cleaning
# ============================================================

import jsonlines

# ── Step 1: Load NEI pairs with metadata ────────────────────
nei_raw = []
with jsonlines.open("nei_pairs_sbert.jsonl") as reader:
    for obj in reader:
        nei_raw.append(obj)

print(f"Loaded: {len(nei_raw)} NEI pairs")

# ── Step 2: Remove high-similarity label-noise pairs ────────
# High-sim pairs with similarity >= 0.80 are likely mislabeled
# as NEI when they actually SUPPORT or REFUTE the claim
def remove_high_sim_noise(pairs, threshold=0.80):
    before = len(pairs)
    cleaned = [
        p for p in pairs
        if not (
            p.get("source") == "bm25_sbert_high"
            and p.get("similarity") is not None
            and p["similarity"] >= threshold
        )
    ]
    print(f"High-sim noise removed: {before - len(cleaned)} pairs (threshold={threshold})")
    return cleaned

nei_denoised = remove_high_sim_noise(nei_raw)

# ── Step 3: Strip metadata, keep only training fields ───────
# Drop source/similarity — not needed for fine-tuning
nei_stripped = [
    {
        "claim":    p["claim"],
        "evidence": p["evidence"],
        "label":    "NOT ENOUGH INFO"
    }
    for p in nei_denoised
]

# ── Step 4: Apply standard cleaning pipeline ────────────────
# Reuses the same apply_cleaning function used for S/R datasets
nei_final = apply_cleaning(nei_stripped, "NEI")

# ── Step 5: Save ─────────────────────────────────────────────
with open("nei_final.jsonl", "w", encoding="utf-8") as f:
    for row in nei_final:
        f.write(json.dumps(row) + "\n")

print(f"Saved: nei_final.jsonl")
print(f"Final NEI count: {len(nei_final)}")
```

```python
import json
# ── Load all three cleaned datasets ─────────────────────────
def load_jsonl(path):
    pairs = []
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            pairs.append(json.loads(line))
    return pairs

supports_final = load_jsonl("supports_final.jsonl")
refutes_final  = load_jsonl("refutes_final.jsonl")
nei_final      = load_jsonl("nei_final.jsonl")

print(f"Loaded SUPPORTS: {len(supports_final)}")
print(f"Loaded REFUTES:  {len(refutes_final)}")
print(f"Loaded NEI:      {len(nei_final)}")
```

```python
import random
# ── Merge ────────────────────────────────────────────────────
merged = supports_final + refutes_final + nei_final

print(f"\nBefore shuffle: {len(merged)} total pairs")

# Shuffle to mix labels evenly — important for training stability
# Without shuffling the model sees all SUPPORTS first, then all REFUTES etc.
random.seed(42)
random.shuffle(merged)

print(f"After shuffle:  {len(merged)} total pairs")

# ── Verify label distribution ────────────────────────────────
from collections import Counter
label_counts = Counter(p["label"] for p in merged)
total = len(merged)

print(f"\n=== Final Label Distribution ===")
for label, count in label_counts.items():
    print(f"  {label:<20} {count:>6} ({count/total*100:.2f}%)")
```

```python
output_path = "final_cleaned_nli_dataset.jsonl"

with open(output_path, "w", encoding="utf-8") as f:
    for row in merged:
        f.write(json.dumps(row) + "\n")

print(f"\nSaved: {output_path}")
print(f"Total: {len(merged)} pairs")
```
